```python
os.chdir(RAW_DATA_PATH)
files = [f.name for f in os.scandir() if f.is_file()]
print(files)

df_p0vsZ = pd.read_csv(files[0])
df_p0vsp1 = pd.read_csv(files[1])
```

```python
df_p0vsZ
```
